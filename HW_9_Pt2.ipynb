```python
#author: Chelsy Gomes

import pandas as pd
import matplotlib.pyplot as plt

cars_data_path = '/content/carsData.csv'
cars_data_df = pd.read_csv(cars_data_path)

# Exercise 1: Create and display a DataFrame using a provided dictionary
exam_data = {'Students': ['Anastasia', 'Dima', 'Katherine', 'Joe', 'Kaely'],
             'Scores': [12.5, 9, 16.5, 14, 12]}
exam_df = pd.DataFrame(exam_data)

# Add a new column called 'Bus Rider' and display the updated DataFrame
bus_rider = ['Y', 'Y', 'N', 'Y', 'N']
exam_df['Bus Rider'] = bus_rider

# Exercise 2: Open 'carsData.csv' file and perform various display and summary operations
# Display the first 5 rows of the cars data
first_five_rows = cars_data_df.head()

# Display summary statistics for the cars data
summary_stats = cars_data_df.describe()

# Display data types in the cars data DataFrame
data_types = cars_data_df.dtypes

# Display the 'acceleration' column (first 5 rows for brevity)
acceleration_column = cars_data_df['acceleration'].head()

# Display summary stats for the 'acceleration' column
acceleration_summary_stats = cars_data_df['acceleration'].describe()

# Sum the values in the 'acceleration' column
acceleration_sum = cars_data_df['acceleration'].sum()

# Exercise 3: Create a separate dataframe containing only 'horse' and 'mpg' columns
horse_mpg_df = cars_data_df[['horse', 'mpg']]

# Replace the index column with 'horse'
horse_mpg_df.set_index('horse', inplace=True)

# Plot a line graph if 'horse' is numeric
if pd.api.types.is_numeric_dtype(horse_mpg_df.index):
    plt.figure(figsize=(10, 5))
    horse_mpg_df['mpg'].plot(title='MPG vs Horsepower')
    plt.ylabel('MPG')
    plt.xlabel('Horsepower')
    plt.show()

# Additional feature: Group by 'cylinder' and calculate the average 'mpg'
cylinder_mpg_avg = cars_data_df.groupby('cylinder')['mpg'].mean()

# Display the outputs
print(exam_df)
print(first_five_rows)
print(summary_stats)
print(data_types)
print(acceleration_column)
print(acceleration_summary_stats)
print(f"Total Acceleration Sum: {acceleration_sum}")
print(horse_mpg_df.head())
print(cylinder_mpg_avg)
```

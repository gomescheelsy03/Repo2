```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Sample data for demonstration
x = [i for i in range(10)]
y = [i**2 for i in x]

# Create a figure and an axis
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

# Plotting the initial graph
line, = ax.plot(x, y, color='blue')

# Adding Slider for interactivity
# Create an axes area for the slider
ax_slider = plt.axes([0.1, 0.1, 0.65, 0.03], facecolor='lightgoldenrodyellow')
# Creating the slider
slider = Slider(ax_slider, 'Exponent', 0.1, 3.0, valinit=2)

# Function to be called when the slider value is changed
def update(val):
    exponent = slider.val
    line.set_ydata([i**exponent for i in x])
    fig.canvas.draw_idle()

# Register the update function with the slider
slider.on_changed(update)

plt.show()
```

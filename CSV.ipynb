```python
import csv

def process_csv(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        data = list(reader)
    return data

def query_data(data):
    def get_choice(change):
        if choice_widget.value == "3":
            print("Exiting...")
            choice_widget.close()
            state_widget.close()
            region_widget.close()
            return
        elif choice_widget.value == "1":
            state_widget.layout.visibility = 'visible'
        elif choice_widget.value == "2":
            region_widget.layout.visibility = 'visible'

    def query_state(change):
        results = [record for record in data if record['State'].lower() == state_widget.value.lower()]
        for r in results:
            print(r)
        choice_widget.value = '0'

    def query_region(change):
        results = [record for record in data if record['Region'].lower() == region_widget.value.lower()]
        for r in results:
            print(r)
        choice_widget.value = '0'

    choice_widget = widgets.Dropdown(
        options={'Select Option': '0', 'State': '1', 'Region': '2', 'Exit': '3'},
        value='0',
        description='Query by:',
    )

    state_widget = widgets.Text(value='', placeholder='Enter state name', description='State:', layout=widgets.Layout(visibility = 'hidden'))
    region_widget = widgets.Text(value='', placeholder='Enter region name', description='Region:', layout=widgets.Layout(visibility = 'hidden'))

    display(choice_widget, state_widget, region_widget)

    choice_widget.observe(get_choice, names='value')
    state_widget.on_submit(query_state)
    region_widget.on_submit(query_region)
```

```python
data = process_csv(filename)
query_data(data)
```

```python
#author: Chelsy Gomes

import matplotlib.pyplot as plt
import numpy as np

# Define the data for the cities and measurements
cities = ['Rome', 'Madrid', 'Istanbul', 'Helsinki']
measurements = {
    'daily_average': [5, 3, 6, 4],
    'highest': [8, 7, 9, 6]
}

# Create an array with the position of each city in the X-axis
x = np.arange(len(cities))

# Define width of the bars
bar_width = 0.35

# Initialize the plot
fig, ax = plt.subplots()

# Plot each set of data
bars1 = ax.bar(x - bar_width/2, measurements['daily_average'], bar_width, label='Daily Average')
bars2 = ax.bar(x + bar_width/2, measurements['highest'], bar_width, label='Highest')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Cities')
ax.set_ylabel('Measurement')
ax.set_title('Daily Average and Highest Temperature Measurements by City')
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()

# Function to attach a text label above each bar
def autolabel(bars):
    """Attach a text label above each bar in *bars*, displaying its height."""
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Call the function to attach the labels
autolabel(bars1)
autolabel(bars2)

# Make the layout more tight and show the plot
fig.tight_layout()
plt.show()
```

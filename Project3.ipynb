```python
#author: Chelsy Gomes
#Due: 10/11/2023
#Project 3
# Always start with option '1' to ensure names are loaded into the program before searching.

# Master Lists for storing names
boy_names = []
girl_names = []

def load_names():
    """Load names from the provided files"""
    global boy_names, girl_names

    # Loading boy names from 'BoyNames.txt' file
    with open('BoyNames.txt', 'r') as f:
        # For each line in the file, strip whitespace and check if the line isn't empty.
        # Add the cleaned line to the boy_names list.
        boy_names = [line.strip() for line in f.readlines() if line.strip()]
    print(f"Loaded {len(boy_names)} boy names.")

    # Loading girl names from 'GirlNames.txt' file
    with open('GirlNames.txt', 'r') as f:
        girl_names = [line.strip() for line in f.readlines() if line.strip()]
    print(f"Loaded {len(girl_names)} girl names.")

def search_names(gender='All', starts_with=''):
    """Search names based on criteria and return matches"""
    # Based on the gender, choose the correct list to search in.
    if gender == 'Boy':
        names = boy_names
    elif gender == 'Girl':
        names = girl_names
    else:
        # If 'All' is chosen, combine both lists.
        names = boy_names + girl_names

    # Filter the list to only include names that start with the given prefix.
    if starts_with:
        names = [name for name in names if name.startswith(starts_with)]

    print(f"Found {len(names)} names based on the criteria.")
    return names

def menu():
    """The main function to present the user menu and handle interactions"""
    while True:
        print("\nMini Search Engine")
        print("1. Load Names")
        print("2. Search Names")
        print("3. Exit")
        choice = input("Enter your choice: ")

        # Handle the user's choice
        if choice == '1':
            load_names()
        elif choice == '2':
            gender = input("Enter Gender (Boy/Girl/All): ").capitalize()
            starts = input("Begin with (or leave blank for no filter): ")
            results = search_names(gender, starts)
            print("\nResults:")
            for name in results:
                print(name)
        elif choice == '3':
            # Exit the program
            break
        else:
            print("Invalid choice!")

# The main entry point of the script
if __name__ == "__main__":
    menu()
```

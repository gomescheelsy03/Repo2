```python
#author: Chelsy Gomes

import matplotlib.pyplot as plt
import numpy as np
cities = ['Rome', 'Madrid', 'Istanbul', 'Houston']

measurements = [6, 1, 3, 5]


plt.figure(figsize=(8, 6))
bars = plt.bar(cities, measurements, color='blue')

# Add a title and labels
plt.title('City Measurements Comparison')
plt.xlabel('City')
plt.ylabel('Measurement')

# Highlight the bar with the highest value - now Rome at 6
bars[0].set_color('green')  # Assuming Rome has the highest measurement with 6
plt.text(0, measurements[0], 'highest', ha='center', va='bottom')

# Calculate the daily average excluding Madrid as it's an outlier
daily_average = (measurements[0] + measurements[2] + measurements[3]) / 3

# Add an annotation for daily average line
plt.axhline(y=daily_average, color='red', linestyle='--')
plt.text(len(cities)-1, daily_average, 'Daily Average', ha='right', va='bottom', color='red')

# Show the plot
plt.show()
```
